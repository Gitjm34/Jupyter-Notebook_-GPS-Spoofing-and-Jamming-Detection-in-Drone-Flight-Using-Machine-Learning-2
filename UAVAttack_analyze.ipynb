```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# üìå CSV ÌååÏùº Í≤ΩÎ°ú (Í≤ΩÎ°ú Ïú†ÏßÄ)
file_path = "C:/Users/user/Desktop/UAVAttackData/Live GPS Spoofing and Jamming/Benign Flight/ace-benign-log_0_2033-8-19-16-27-30_vehicle_gps_position_0.csv"

# üìå Îç∞Ïù¥ÌÑ∞ Î°úÎìú
df = pd.read_csv(file_path)

# üìå Î∂ÑÏÑùÌï† Ï£ºÏöî Ïª¨Îüº ÏÑ†ÌÉù
features = ["timestamp", "lat", "lon", "alt", "alt_ellipsoid", "eph", "epv",
            "hdop", "vdop", "noise_per_ms", "jamming_indicator", "vel_m_s",
            "vel_n_m_s", "vel_e_m_s", "vel_d_m_s", "heading", "fix_type", "satellites_used"]

df_selected = df[features]

# üìå 1. Í∏∞Î≥∏ ÌÜµÍ≥Ñ Î∂ÑÏÑù
print("üìå Í∏∞Î≥∏ ÌÜµÍ≥Ñ Î∂ÑÏÑù")
print(df_selected.describe())

# üìå 2. Ïù¥ÏÉÅÏπò ÌÉêÏßÄ (IQR Î∞©Ïãù)
Q1 = df_selected.quantile(0.25)
Q3 = df_selected.quantile(0.75)
IQR = Q3 - Q1
outliers = ((df_selected < (Q1 - 1.5 * IQR)) | (df_selected > (Q3 + 1.5 * IQR))).sum()

print("\nüìå Ïù¥ÏÉÅÏπò Í∞úÏàò (IQR Í∏∞Ï§Ä)")
print(outliers)

# üìå 3. Î∂ÑÌè¨ ÏãúÍ∞ÅÌôî (ÌûàÏä§ÌÜ†Í∑∏Îû® & KDE)
plt.figure(figsize=(12, 8))
for i, col in enumerate(features[1:]):  # timestamp Ï†úÏô∏
    plt.subplot(4, 5, i+1)
    sns.histplot(df_selected[col], kde=True, bins=30)
    plt.title(col)
plt.tight_layout()
plt.show()

# üìå 4. GPS Í¥ÄÎ†® Feature ÎπÑÍµê (Noise, Jamming, Accuracy)
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_selected[["noise_per_ms", "jamming_indicator", "eph", "epv"]])
plt.title("üìå GPS Ïã†Ìò∏ ÌíàÏßà Î∞è ÍµêÎûÄ ÏßÄÌëú ÎπÑÍµê")
plt.grid(True)
plt.show()

# üìå 5. ÏÉÅÍ¥Ä Í¥ÄÍ≥Ñ Î∂ÑÏÑù (Heatmap)
plt.figure(figsize=(10, 8))
sns.heatmap(df_selected.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("üìå Feature Correlation Matrix")
plt.show()
```
